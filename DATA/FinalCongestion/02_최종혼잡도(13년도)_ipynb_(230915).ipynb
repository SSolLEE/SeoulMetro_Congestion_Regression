### 파일 불러오기

```python
import pandas as pd

mons = pd.read_csv("/content/0915_역별월별비율_15_22.csv")
days = pd.read_csv("/content/0915_역별요일비율(결측치해결)_인천7빼고.csv")
```

```python
# 엑셀 파일에서 모든 시트를 읽어서 딕셔너리 형태로 저장
excel_file = pd.ExcelFile('/content/2013년 혼잡도.xlsx')
sheet_names = excel_file.sheet_names  # 시트 이름 목록
sheet_names
```

### 엑셀파일 읽기 및 데이터프레임으로 만들기

```python
# 모든 시트의 데이터를 저장할 빈 리스트를 생성
all_dfs = []

for sheet_name in sheet_names:
    # 필요한 행부터 읽기 시작
    df = pd.read_excel(excel_file, sheet_name=sheet_name, header=2)

    # 불필요한 열 삭제
    df.dropna(axis=1, how='all', inplace=True)

    # 요일구분 컬럼 추가
    day_of_week = sheet_name.split()[-2]  # 시트 이름에서 요일 정보 추출
    df['요일구분'] = day_of_week

    # 상하구분 컬럼 추가 (상선 또는 하선)
    direction = '상선' if '상선' in sheet_name else '하선'
    df['상하구분'] = direction

    # 데이터프레임을 리스트에 추가
    all_dfs.append(df)
```

```python
all_dfs
```

```python
# 리스트에 있는 모든 데이터프레임을 합치기
df_13 = pd.concat(all_dfs, ignore_index=True)
```

### 결측값 처리 (호선명 채우기)
- 엑셀에서 읽어와서 공통된 셀에 포함된 경우 데이터프레임 처리할 때 특정 열의 값이 빈 채로 출력됨.
- 호선값이 비어있을 경우 직전 행의 호선 값을 복사해서 결측 처리

```python
# 비어있는 호선 값 채우기
for index, row in df_13.iterrows():
    if pd.isnull(row['호선']):  # 호선 값이 NULL인 경우
        df_13.at[index, '호선'] = df_13.at[index - 1, '호선']  # 직전 행의 호선 값 복사
```

### df_13 데이터 형식 수정

```python
# 호선 문자열 수정
df_13['호선'] = df_13['호선'].str.extract('(\d+)').astype(int)

# 시간대 실수형으로 형식 바꾸기
time_columns = ['05시~06시', '06시~07시', '07시~08시', '08시~09시', '09시~10시',
                '10시~11시', '11시~12시', '12시~13시', '13시~14시', '14시~15시',
                '15시~16시', '16시~17시', '17시~18시', '18시~19시', '19시~20시',
                '20시~21시', '21시~22시', '22시~23시', '23시~24시', '24시~01시']

for col in time_columns:
    df_13[col] = df_13[col].astype('float64')
```

### df_13 역 이름 수정  
- 다른 연도의 지하철 역 이름과 통일


```python
# 을지입구, 을지3가, 을지4가 -> 을지로입구, 을지로3가, 을지로4가 변경
df_13['역명'] = df_13['역명'].replace('을지입구', '을지로입구')
df_13['역명'] = df_13['역명'].replace('을지3가', '을지로3가')
df_13['역명'] = df_13['역명'].replace('을지4가', '을지로4가')

# 성내 -> 잠실나루 변경
df_13['역명'] = df_13['역명'].replace('성내', '잠실나루')

# 이대입구 -> 이대 변경
df_13['역명'] = df_13['역명'].replace('이대입구', '이대')

# 성신여대 -> 성신여대입구 변경
df_13['역명'] = df_13['역명'].replace('성신여대', '성신여대입구')

# 신천 -> 잠실새내 변경
df_13['역명'] = df_13['역명'].replace('신천', '잠실새내')
```

- 역이름 뒤에 () 들어간 내용 삭제

```python
# 역 이름 뒤 괄호와 괄호 안 내용 삭제
df_13['역명'] = df_13['역명'].str.replace(r'\(.*\)', '', regex=True)
```
